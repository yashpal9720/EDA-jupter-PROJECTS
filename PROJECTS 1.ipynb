```python
import pandas as pd 
import matplotlib.pyplot as plt         
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head)
```

```python
# change weekly study hours column 
df["WklyStudyHours"] = df["WklyStudyHours"] .str.replace(" 5 - 10", " 5 - 10")
df.head()
```

```python
# Gender Distribution 
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show
```

```python
# from the above chart we have analysed that:
# the number of females in the data is more than the number of males 
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gp)
```

```python
sns.heatmap(gp, annot=True)
plt.title("Gender Distribution")
plt.figure(figsize=(4,4))
plt.show
```

```python
# from that above chart we have concluded that the education of the parents have a good impact 
```

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gp1)
```

```python
sns.heatmap(gp1, annot=True)
plt.title("Relation between Parent's Education and student score ")
plt.figure(figsize=(4,4))
plt.show
```

```python
# from the above chart we have concluded that there is no negligible impcat on the 
# student' score due to their parent's marital status 
```

```python
import matplotlib.pyplot as plt

# Count the number of rows for each ethnic group
groupA = df.loc[df['EthnicGroup'] == "group A"].shape[0]
groupB = df.loc[df['EthnicGroup'] == "group B"].shape[0]
groupC = df.loc[df['EthnicGroup'] == "group C"].shape[0]
groupD = df.loc[df['EthnicGroup'] == "group D"].shape[0]
groupE = df.loc[df['EthnicGroup'] == "group E"].shape[0]

# Create a list of counts
mlist = [groupA, groupB, groupC, groupD, groupE]
print(mlist)

# Plot the pie chart
plt.pie(mlist, labels=["Group A", "Group B", "Group C", "Group D", "Group E"], autopct='%1.1f%%')
plt.title('Distribution of Ethnic Groups')
plt.show()
```
